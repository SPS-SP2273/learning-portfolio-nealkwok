You are a member of an international team analysing environmental pollution. The project involves ten cities: Cairo, Dhaka, Jakarta, Karachi, Manila, Melbourne, Osaka, Shanghai, Singapore, and Tokyo.

Your task is to process and organise data from various environmental measurements. The provided zip file, os-collaboration-exercise-data.zip, contains all relevant data files for the past year.

Things to note

1. Data Files: Each data file is named in the format `month-date_city.txt` (e.g., `may-10_singapore.txt`).

## Task 1

We like to convert the filename to a more useful numerical format. Specifically, we want to convert `month-date-city.txt` (i.e. `may-10-singapore.txt`) to a `mm-dd-city.txt` (i.e. `05-10-singapore.txt`) format.

Using the following dictionary (or otherwise), write a snippet of Python code to convert '`oct-08_singapore.txt`' to `10-08-singapore.txt`.

```python
# name the library :"D
month_number_dictionary = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

# give filename
# not a function, don't give filename a value AFTER the following parts since python runs in SEQUENCE and since we aren't using a function
filename = 'oct-08_singapore.txt'

parts = filename.split('_')
month, day = parts[0].split('-')
city = parts[1].split('.')[0]
month = month_number_dictionary[month.lower()]

print(f"{month}-{day}-{city}.txt")
```

## Task 2

Incorporate your previous code into a function named `rename_my_file(old_file_name)` that accepts the old filename as the argument and returns the new file name.

```python
month_number_dictionary = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

# incorporating previous code into a function
def rename_file_name(old_file_name):
    parts = old_file_name.split('_')
    month, day = parts[0].split('-')
    city = parts[1].split('.')[0]
    month = month_number_dictionary[month.lower()]
    return f"{month}-{day}-{city}.txt"

# reusing previous task example for an output
old_file_name = 'oct-08_singapore.txt'
new_file_name = rename_file_name(old_file_name)
print(new_file_name)  
```

## Task 3

```python
# provided file list
provided_files_list = [
    'oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
    'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
    'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
    'jan-01_tokyo.txt'
]

# Rename files using the function
for old_file_name in provided_files_list:
    new_file_name = rename_file_name(old_file_name)
    print(f"{old_file_name} ----> {new_file_name}")
```

## Task 4

```python
import glob
import os

folder_path = "/Users/nealkwok/Documents/neal_pc/nus/academics/sps_special programme in science/sp2273/learning-portfolio-nealkwok/files, folders & os  exercises/os-collaboration-exercise-data"

txt_files = glob.glob(os.path.join(folder_path, '*.txt'))

for file in txt_files:
    old_file_name = file
    new_file_name = os.path.join(folder_path, rename_file_name(os.path.basename(file)))
    os.rename(old_file_name, new_file_name)
    print(f"Renamed '{old_file_name}' to '{new_file_name}'")
```

## Task 5

Use a `for` loop to create a folder for each city.

```python
os.mkdir('city')

for destination in ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']:
    path = os.path.join('city', destination)
    try:
        os.mkdir(path)
        print(f'Creating {path}')
    except FileExistsError:
        print(f'{path} already exists; skipping creation.')
```
